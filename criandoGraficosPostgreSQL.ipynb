```python
import psycopg2
import matplotlib.pyplot as plt
#O comando abaixo não permite que o gráfico abra em outro guia
%matplotlib notebook
print('Criando conexão')
conn = psycopg2.connect(
    host='localhost',
    user='postgres',
    password='postgres',
    database='postgres')

#criando um cursor
c = conn.cursor()
```

```python
def criaGrafico():
    c.execute("select id, valor from produtos")
    ids= []
    valores = []
    dados = c.fetchall()
    for items in dados:
        ids.append(items[0])
        valores.append(items[1])
        
    plt.bar(ids,valores)
    plt.show()
```

```python
criaGrafico()
```
